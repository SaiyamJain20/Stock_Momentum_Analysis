```python
import numpy as np
import pandas as pd
import requests
import math
from scipy import stats
import yfinance as yf
```

```python
stocks = pd.read_csv('sp_500_stocks.csv')
stocks
```

```python
from time import sleep

final_data = []
hqm_data = []

for ticker in stocks['Ticker']:
    try:
        stock = yf.Ticker(ticker)
        info = stock.info

        price = info.get('currentPrice')
        market_cap = info.get('marketCap')

        final_data.append({
            'Ticker': ticker,
            'Price': price,
            'Market Capitalization': market_cap,
            'Number Of Shares to Buy': 'N/A'
        })

        stock_prices = yf.download(ticker, period='50y', interval='1d', auto_adjust=True)
        # print(stock_prices)

        close_prices = stock_prices['Close'][ticker].dropna()

        one_year_return = (close_prices.iloc[-1] / close_prices.iloc[-252]) - 1 if len(close_prices) >= 252 else (close_prices.iloc[-1] / close_prices.iloc[-len(close_prices)]) - 1
        six_month_return = (close_prices.iloc[-1] / close_prices.iloc[-126]) - 1 if len(close_prices) >= 126 else (close_prices.iloc[-1] / close_prices.iloc[-len(close_prices)]) - 1
        three_month_return = (close_prices.iloc[-1] / close_prices.iloc[-63]) - 1 if len(close_prices) >= 63 else (close_prices.iloc[-1] / close_prices.iloc[-len(close_prices)]) - 1
        one_month_return = (close_prices.iloc[-1] / close_prices.iloc[-21]) - 1 if len(close_prices) >= 21 else (close_prices.iloc[-1] / close_prices.iloc[-len(close_prices)]) - 1


        # print(f"Ticker: {ticker}")
        # print(f"One Year Return: {one_year_return}")
        # print(f"Six Month Return: {six_month_return}")
        # print(f"Three Month Return: {three_month_return}")
        # print(f"One Month Return: {one_month_return}")

        hqm_data.append({
            'Ticker': ticker,
            'Price': price,
            'Number of Shares to Buy': 'N/A',
            'One-Year Price Return': one_year_return,
            'One-Year Return Percentile': 'N/A',
            'Six-Month Price Return': six_month_return,
            'Six-Month Return Percentile': 'N/A',
            'Three-Month Price Return': three_month_return,
            'Three-Month Return Percentile': 'N/A',
            'One-Month Price Return': one_month_return,
            'One-Month Return Percentile': 'N/A',
            'HQM Score': 'N/A'
        })

        sleep(0.1)

    except Exception as e:
        print(f"Failed for {ticker}: {e}")

final_dataframe1 = pd.DataFrame(final_data)
hqm_dataframe = pd.DataFrame(hqm_data)

final_dataframe1
hqm_dataframe
```

```python
portfolio_size = input("Enter the value of your portfolio:")

try:
    val = float(portfolio_size)
except ValueError:
    print("That's not a number! \n Try again:")
    portfolio_size = input("Enter the value of your portfolio:")
```

```python
position_size = float(portfolio_size) / len(final_dataframe1.index)
for i in range(len(final_dataframe1)):
    price = final_dataframe1.loc[i, 'Price']

    if pd.notnull(price) and price > 0:
        final_dataframe1.loc[i, 'Number Of Shares to Buy'] = math.floor(position_size / price)
    else:
        final_dataframe1.loc[i, 'Number Of Shares to Buy'] = 0

final_dataframe1
```

```python
time_periods = [
                'One-Year',
                'Six-Month',
                'Three-Month',
                'One-Month'
                ]

for row in hqm_dataframe.index:
    for time_period in time_periods:
        hqm_dataframe.loc[row, f'{time_period} Return Percentile'] = stats.percentileofscore(hqm_dataframe[f'{time_period} Price Return'], hqm_dataframe.loc[row, f'{time_period} Price Return'])/100

for time_period in time_periods:
    print(hqm_dataframe[f'{time_period} Return Percentile'])

hqm_dataframe
```

```python
from statistics import mean

for row in hqm_dataframe.index:
    momentum_percentiles = []
    for time_period in time_periods:
        momentum_percentiles.append(hqm_dataframe.loc[row, f'{time_period} Return Percentile'])
    hqm_dataframe.loc[row, 'HQM Score'] = mean(momentum_percentiles)
```

```python
hqm_dataframe.sort_values(by = 'HQM Score', ascending = False)
hqm_dataframe = hqm_dataframe[:51]
```
